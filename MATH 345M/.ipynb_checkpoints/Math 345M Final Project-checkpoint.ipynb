First, though, lets get the seven number summary of this data and breaking down based on that statistical information.

```python
def pandas_seven_num_sum(df):
    # Compute the mean first
    mean = df.mean()
    
    # Now the standard deviation
    std = df.std()
    
    # Now we can move on to our five number summary, starting with our min:
    min_val = df.min()
    
    # Now the first quartile. Since I am working with pandas now I have a different method for doing this than with numpy.
    firstq = df.quantile(.25)
    
    # Now the third quartile
    thirdq = df.quantile(.75)
    
    # Now for the median:
    median = df.median()
    
    # And finally for the max
    max_val = df.max()
    
    # Now we just handle the printing logic
    print("The mean is:", mean)
    print("The standard deviation is:", std)
    print("The median is:", median)
    print("The min is:", min_val)
    print("The max is:", max_val)
    print("The first quartile is:", firstq)
    print("The third quartile is:", thirdq)
    
    return mean, std, median, min_val, max_val, firstq, thirdq
```

```python
mean, std, median, min_val, max_val, firstq, thirdq = pandas_seven_num_sum(top_speed)
```

So now we have some statistical information about the Top Speed data set. We know we have an average top speed of about 185 mph with typical deviations of about 34 mph. Now that we have this statistical information, lets go ahead and try to pinpoint some outliers in our data set using two different methods- The empirical test and the iqr test.

### Empirical Test

The structure for performing the Empirical test is as follows:

1.   Find mean and std of data set.
2.   Multiply the standard deviation by 3.
3.   Add $3*std$ to the mean to find the high cutoff.
4.   Subtract $3*std$ from the mean to find the low cutoff.

We already have our mean and standard deviation, so we have everything we need to start calculating!

```python
# Establish our low cutoff
low_cutoff = mean - 3 * std

# Establish our high cutoff
high_cutoff = mean + 3 * std

# Populate a list with outliers
outliers = [i for i in top_speed if i > high_cutoff or i < low_cutoff]
print(f"outliers: {outliers}")
```

So according to our Empirical test this particular data set has no outliers, everything falls within 3 standard deviations of the mean. That said, I don't necessarily believe that the Empirical test is our best option for this data set, as the histogram displays that the data isn't quite guassian, but rather looks a bit skewed from my perspective. The IQR test, then, should be a better outlier identification procedure for this data set, as that is better for sets that aren't quite gaussian.

### IQR Test

The structure for the IQR test is as follows:

1.   Calculate the interquartile range for the data.
2.   Multiply the interquartile range ($IQR$) by $1.5$
3.   Add $1.5 (IQR)$ to the third quartile. Any number greater than this is a suspected outlier.
4.   Subtract $1.5  (IQR)$ from the first quartile. Any number less than this is a suspected outlier.

We already have our first and third quartiles, so we can begin caclulating!

```python
iqr = thirdq - firstq # IQR is the third quartile - first quartile

high_cutoff = (thirdq + (1.5*iqr)) # Find the high cutoff by adding the third quartile to 1.5*IQR. Values greater
                                               # than the high cutoff will be considered outliers.
    
low_cutoff = (firstq - (1.5*iqr)) # Find the low cutoff by subtracting 1.5*IQR from the first quartile. Values lower than
                                      # than the low cutoff will be considered outliers.
outliers = [i for i in top_speed if i > high_cutoff or i < low_cutoff]
good_data = [i for i in top_speed if i <= high_cutoff and i >= low_cutoff]

print(f"Outliers: {outliers}")
```

So according to our IQR test, this data set does have 1 outlier- a 285 mph top speed. So at least we know that our data is pretty reasonable overall, there is just one value that is somewhat questionable. I don't believe this single outlier is worth removing, as I think it will add valuable information to our spread of speeds and removing it will damage the viability of our results.

We can also take a look at how our data would change if we removed this outlying data point, as represented by the histogram below:

```python
plt.figure(figsize = (12,12))
plt.hist(good_data, bins=50, color='b')
plt.title("Top Speed Historgram\n", fontsize=40)

plt.xlabel("\nTop Speed (mph)", fontsize=30)

plt.tick_params(direction='out', length=10)

plt.xticks(fontsize=25, rotation=90)
plt.yticks(fontsize=25, rotation=0)

plt.show()
```

Removing that one outlier doesn't really seem to impact the shape of our graph all that much, meaning that removing it doesn't really add anything to our data set. Thus, I move to keep the outlier in my data set, which means that now we can move on to my hypothesis testing!

So we find that we have two differently sized samples and, on top of that, we don't know anything really about the larger population of sports cars. This means that if we want to compare these sets of speeds, the only option that is really available to us would be to use a students two-sample t-test and compare the averages from each sample.

Lets assume we have two population means $\mu_{est}$ and $\mu_{new}$ that describe the populations associated with these samples. Based on my understanding of cars, I am inclined to believe that these cars that are fairly well established will indeed be faster than their newer cousins. This will be our alternative hypothesis, since our null hypothesis is a statement of no difference between the population means. Thus, we have

$H_0 : \mu_{est} = \mu_{new}$

$H_a: \mu_{est} > \mu_{new}$

```python
# Variable declaration

established_n = len(established_speeds)
established_avg = np.mean(established_speeds)
established_std = np.std(established_speeds, ddof = 1)

new_n = len(new_speeds)
new_avg = np.mean(new_speeds)
new_std = np.std(new_speeds, ddof = 1)
```

Since we will be working with a two sample t-distribution, we are going to need compute our degrees of freedom for the two lists. Typically, we compute the degrees of freedom as 

$\huge{df = {({{(s_1)^2 \over n_1} + {(s_2)^2 \over n_2}})^2 \over {({1 \over {n_1 - 1}})({(s_1)^2 \over n_1})^2 + ({1 \over {n_2 - 1}})({(s_2)^2 \over n_2})^2}}}$

where $s_1$ and $s_2$ are the standard deviations of our samples and $n_1$ and $n_2$ are our sample sizes.

When we were working on Homework 6 I wrote a function that replicated this computation for me which is included below:

```python
def find_deg_freedom(s1, s2, n1, n2):
    """
    Finds the degrees of freedom for two lists
    s1 is the standard deviation of list1
    s2 is the standard deviation of list2
    n1 is the length of list1
    n2 is the length of list2
    """
    # Breaking the equation into bits so I don't have to write a single nasty arithmetic operation
    bit1 = (s1**2)/n1
    bit2 = (s2**2)/n2
    
    bit3 = 1/(n1-1)
    bit4 = bit1**2
    
    bit5 = 1/(n2-1)
    bit6 = bit2**2
    
    df = ((bit1 + bit2)**2)/((bit3)*(bit4) + (bit5)*(bit6))
    
    return df
```

```python
dof = find_deg_freedom(established_std, new_std, established_n, new_n)
print("Degrees of Freedom =", dof)
```

So, we now know that we will be performing a two-sample t-test with about 71 degrees of freedom.

Further, since we are using a two-sample t-test to compare two sample means, we know we will have a random variable $\bar X_{est} - \bar X_{new}$ that represents the mean difference between the top speeds of established and newer car models.

In order to find our p-value for this situation we are going to need to find our t-score for this test. The formula for the t-score is as follows:

$\huge{t = {{(\bar x_1 - \bar x_2) - (\mu_1 - \mu_2)} \over {\sqrt{({(s_1)^2 \over n_1}) + ({(s_2)^2 \over n_2})}}}}$

Where $\bar x_1$ and $\bar x_2$ are the sample means, $\mu_1 - \mu_2$ is the *expected* difference between the population means based on the null hypothesis, $n_1$ and $n_2$ are the sample sizes, and $s_1$ and $s_2$ are the sample standard deviations.

I wrote another function on Homework 6 to do this process for me, and since from our null hypothesis we assume that there is no difference in the mean top speeds, we want $\mu_1 - \mu_2 = 0$, meaning $\mu_1 = \mu_2$.

```python
def find_2Samp_test_stat(xbar1, xbar2, s1, s2, n1, n2, mu1, mu2):
    """
    Finds the test statistic of a two sample t test
    xbar1 is the mean of the first sample
    xbar2 is the mean of the second sample
    s1 is the standard dev of the first sample
    s2 is the standard dev of the second sample
    n1 is the size of the first sample
    n2 is the size of the second sample
    mu1 is the population mean associated with the first sample
    mu2 is the population mean associated with the second sample
    """
    # Once again breaking the equation into bits to ensure that I don't make a mistake
    bit1 = (s1**2)/n1
    bit2 = (s2**2)/n2
    
    numer = (xbar1 - xbar2) - (mu1 - mu2)
    denom = (bit1 + bit2)**(1/2)
    
    t_score = numer/denom
    return t_score
```

```python
t_score = find_2Samp_test_stat(established_avg, new_avg, established_std, new_std, established_n, new_n, 0, 0)
print("T-Score =", t_score)
```

Now that we have our t-score, we can move on to calculating our p-value and determining if we should be rejecting or keeping our null hypothesis. We can use the Survival Function from the t-distribution section of scipy.stats in order to calculate this p-value:

```python
p_value = sp.t.sf(abs(t_score), dof)
print("P-Value =", p_value)
```

My understanding of what this p-value means has always been a little weird for me. From my understanding, the p-value represents the likelihood that, assuming we find that our null hypothesis should not be rejected, we can replicate those same results with a different random sample. So in this situation, if we find that our null hypothesis should not be rejected, there is only about a 1.68% chance of actually replicating those (or more extreme) results using a different sample from this population of cars. So, if our null hypothesis does turn out to be unable to be rejected, it seems there is only about a 1.68% chance that we can find another random sample for which the null hypothesis is still non-rejectable, implying that our original sample represented outlying data.

We can take a look at what values fall within our range of p-values by graphing this situation. It is important to note that this is a non-standardized graph, meaning that the units of our x-range are in miles per hour and represent the spread of mean differences. Our p-value is extremely tiny, so the range of values it covers will be pretty difficult to see on a typical graph, as we can see below:

```python
avg = established_avg - new_avg

x = np.arange(abs(avg), 15, 0.01)
x_all = np.arange(-15, 15, 0.01)

y = sp.t.pdf(x, dof)
y_all = sp.t.pdf(x_all, dof)

fig, ax = plt.subplots(figsize = (10, 7))
plt.style.use('fivethirtyeight')
ax.plot(x_all, y_all)

ax.fill_between(x, y, 0, alpha = 0.3, color = 'b')
ax.fill_between(x_all, y_all, 0, alpha = 0.1)

plt.scatter(abs(avg), 0, s=100)
plt.annotate("P-value = 0.01684258074218584", (abs(avg), 0))

ax.set_xlim([-15, 15])
ax.set_xlabel('Mean Differences (mph)')
ax.set_yticklabels([])
plt.title('Mean Difference in Established vs New Car Top Speeds')
plt.show()
```

In order to actually see the p-region, I zoomed in to the graph to only display the x range from the abs(mean) to 15, which was the end of my x range. We can see just how ridiculously small this p-value is by the fact that the only way to display this shaded portion is to set our y limit to 0.0000000000000000001, so what is being displayed in the graph below has a y-height ranging just from [0, 0.0000000000000000001].

```python
avg = established_avg - new_avg

x = np.arange(abs(avg), 15, 0.01)
x_all = np.arange(-15, 15, 0.01)

y = sp.t.pdf(x, dof)
y_all = sp.t.pdf(x_all, dof)

fig, ax = plt.subplots(figsize = (10, 7))
plt.style.use('fivethirtyeight')
ax.plot(x_all, y_all)

ax.fill_between(x, y, 0, alpha = 0.3, color = 'b')
ax.fill_between(x_all, y_all, 0, alpha = 0.1)

ax.set_xlim([abs(avg)-0.5, 15])
ax.set_ylim([0, 0.0000000000000000001])

plt.scatter(abs(avg), 0, s=100)
plt.annotate("P-value = 0.01684258074218584", (abs(avg), 0))

plt.scatter(15, 0.0000000000000000001, s=100)
plt.annotate("y = 0.0000000000000000001", (15, 0.0000000000000000001))

ax.set_xlabel('Mean Differences (mph)')
ax.set_yticklabels([])
plt.title('Mean Difference in Established vs New Car Top Speeds')
plt.show()
```

Now we can move on to the big test, determining whether we should reject our null hypothesis or not. I propose that an alpha of 0.05 will be more than enough as we are not dealing with a life or death situation here.

```python
def rejectCheck(alpha, p):
    # Checks if a null hypothesis should be rejected. A return of 'True' indicates H_0 should be rejected.
    # A return of 'False' indicates H_0 should not be rejected.
    return alpha > p

alpha = 0.05
rejectCheck(alpha, p_value)
```

So, based on the outcome of my totally necessary and not at all pointless reject check function, we find that we should reject our null hypothesis in favor of our alternative hypothesis. What this tells us is that at a significance level of 5%, we have sufficient evidence to reject the claim that there is no difference in average top speeds for established cars vs new cars. This leads us to accept our alternative hypothesis as true, that the mean top speeds of established cars is strictly greater than the mean top speeds of newer cars.

However, one thing that is worth noting is that if we instead use a significance level of 0.01, we will find that we *should not* reject our null hypothesis. So while we do have sufficient evidence at the 5% level of siginificance, we do *not* have enough evidence to support the rejecting of our null hypothesis if we instead test at a 1% significance level. This doesn't mean that we would suppose that our null hypothesis is true in that situation, just that we know we can't say that it is definitely false. Given this, I would argue that perhaps this area needs more research in order to establish if we can reliably reject our null hypothesis at the 1% level of significance.

```python
weight_frame = df["Weight(kg)"]
hp_frame = df["Horsepower"]
acc_frame = df["0-60 Time(sec)"]
```

### Horsepower vs Acceleration Times

I want to start my testing by taking a look at how horsepower compares to acceleartion. The first step for any linear regression test is to construct a scatter plot of the data, so I'm going to start with that.

```python
plt.scatter(hp_frame, acc_frame, marker = 'o')
plt.title("Car 0-60 times compared to horsepower")
plt.xlabel("Horsepower")
plt.ylabel("0-60 time (sec)")
```

One thing we can notice just using eyesight here is that there seems to be more of an exponential relationship with our acceleration times and our horsepower, which suggests that maybe using linear regression isn't going to give us the best fit possible here. This might be a problem for us because we were going to use the `sp.linregress` method in order to construct our regression line, but that won't be as effective for a non-linear graph and we don't have any easy functions like `sp.linregress` for exponential fits, and we specifically want to study this data by using the linear regression techniques we learned in class. One thing that we could do here is scale our data to a log-log graph, and then using the `sp.linregress` method on that log-scaled data will return a line that represents a logarithmic curve, as the context of the graph will already be logarithmic, and a line in a log-log graph represents an exponential curve in a typical x-y plane. 

All this requires us to do is find the logarithmic values of our x and y data sets and then plot those on a scatter plot, and based on our curve above we should see something a bit more linear.

```python
hp_ln = np.array([math.log(i) for i in hp_frame])
acc_ln = np.array([math.log(i) for i in acc_frame])
plt.scatter(hp_ln, acc_ln, marker = 'o')
plt.title("ln(0-60 times compared to ln(horsepower)")
plt.xlabel("ln(Horsepower)")
plt.ylabel("ln(0-60 time (sec))")
```

Okay, so now we have some data that is just a bit more linear, which is good for us because now our linear regression techniques will be a bit more applicable, we just have to remember that we are looking at a log-log graph at the moment and not a typical x-y plane. This change in our data shouldn't impact the reliability of our linear regression test, we have just changed what aspect of the data we are looking at. If we now discover a significant linear correlation between the log(0-60 times) and log(horsepower), that implies a significant *exponential* correlation between 0-60 times and horsepower due to how log-log graphs change the way the data is displayed.

Our next step is to collect the information about our linear regression line and then use that data to construct the equation of our line. 

```python
hslope, hint, hrval, hpval, hstderr = sp.linregress(hp_ln, acc_ln)
print("Slope:", hslope)
print("Intercept:", hint)
print("R-value:", hrval)
print("P-value:", hpval)
print("Standard Error:", hstderr)
```

Okay! Now we are armed with enough information to construct our linear regression line for this log-log graph! Using the following variable definitions:

1. x = hp_frame[i]

2. y = acc_frame[i]

Then the data we collected above corresponds to the following linear function (note that `math.log()` finds the natural log of the argument if no base is specified):

$ln(y) = -0.416*ln(x) + 3.85$

From this equation we can determine the follwing information:

We have slope of about -0.416, which tells us that as the natural log of horsepower goes up by a factor of 1, our associated natural log of acceleration rate decreases by a factore of -0.416. Our intercept tells us that when the natural log of horsepower is 0, the associated natural log of acceleration rate is about 3.85 seconds.

Once again, though, we are going to be looking at the significance of the correlation on our log-log graph as we want to use our linear regression techniques to perform this test. Let's overlay our line equation over the scatter plot data and take a peek at how it seems to correlate.

```python
plt.scatter(hp_ln, acc_ln, marker = 'o')
plt.plot(hp_ln, hslope*hp_ln + hint)
plt.title("ln(0-60 times compared to ln(horsepower)")
plt.xlabel("ln(Horsepower)")
plt.ylabel("ln(0-60 time (sec))")
```

With all this information we have, we can now begin the process of determining the strength of this correlation. Just based off of eyesight it seems like there is a pretty good correlation here, but our numbers never lie.

From the `sp.linregress` method we alread know that R = -0.784, but now we need to determine if that correlation is significant. We use the value $\rho$ to represent our ln(population correlation coefficient), which is unknown. If $\rho$ = 0, then we can determine that there is no significant linear relationshipt between ln(x) and ln(y). However, if $\rho \ne 0$, then there IS some significant linear relationship between ln(x) and ln(y). Based on the information in our book, we should have the following null and alternative hypotheses:

$H_0: \rho = 0$

A.K.A: $H_0:$ There is not a significant linear relationship 

$H_a: \rho \ne 0$

A.K.A: $H_a:$ Ther is a significant linear relationship

We already have that P = $8.37 * 10^{-27}$ from our `sp.linregress` function, so all we have left to do is motivate a value for alpha and compare the two. Given how crazy tiny our p-value is, we can already tell that we are going to be able to reject our null hypothesis, but we should still perform the test to be sure. This isn't a matter of life and death, so I think that an alpha of 0.05 should be suitable.

```python
alpha = 0.05

rejectCheck(alpha, hpval)
```

Okay! So we find that we should reject the null hypothesis claim that there is no significant linear correlation between the natural log of our horsepower values and the natural log of our acceleration values. This leads us to accept our alternative hypothesis as true, that there is a significant linear correlation between the natural log of the horsepower values and the natural log of the acceleration values. This further tells us that, in terms of our original x-y plane, we find we have a significant exponential correlation between the horsepower values and the acceleration times. 

However, just establishing that this correlation is significant isn't enough to finish this particular analysis. Now we need to perform the same process for our weight to acceleration times and verify if that test has a significant correlation as well, and, if it does, which of our two hypothesis tests imply the greatest correlation.

### Weight vs Acceleration Times

Since in the last part we scaled all of our data to a log-log graph we will have to do the same in this situation to maintain consistency, but I still want to take a look at the non-scaled graph to see what that data looks like.

```python
plt.scatter(weight_frame, acc_frame, marker = 'o')
plt.title("Car 0-60 times compared to weight")
plt.xlabel("Weight (kg)")
plt.ylabel("0-60 time (sec)")
```

So in this situation the data is a bit all over the place, there certainly isn't quite as much of an obvious correlation as the Horsepower vs Acceleration Times had, but lets once again scale this data to a log-log graph:

```python
weight_ln = np.array([math.log(i) for i in weight_frame])

plt.scatter(weight_ln, acc_ln, marker = 'o')
plt.title("ln(0-60 times) compared to ln(weight)")
plt.xlabel("ln(Weight (kg))")
plt.ylabel("ln(0-60 time (sec))")
```

That certainly didn't do our data any favors, it almost seems to be more spread out now. However, we have to map this data to a log-log graph so we can actually compare our resulting R values, not doing so would remove our ability to accurately compare our outcomes. 

Our next step is to generate the values we need to create our least squares regression line:

```python
wslope, wint, wrval, wpval, wstderr = sp.linregress(weight_ln, acc_ln)
print("Slope:", wslope)
print("Intercept:", wint)
print("R-value:", wrval)
print("P-value:", wpval)
print("Standard Error:", wstderr)
```

Okay! Now we are armed with enough information to construct our linear regression line for this log-log graph! Using the following variable definitions:

1. x = weight_frame[i]

2. y = acc_frame[i]

Then the data we collected above corresponds to the following linear function (note that `math.log()` finds the natural log of the argument if no base is specified):

$ln(y) = -0.063*ln(x) + 1.75$

From this equation we can determine the follwing information:

We have slope of about -0.063, which tells us that as the natural log of weight goes up by a factor of 1, our associated natural log of acceleration rate decreases by a factor of -0.063, which is an odd correlation as it suggests that, at least in this data set, we see a decrease in acceleration time when we have an increase in weight. I'm going to go a bit more into depth about why I think this the case after we finish up the hypothesis testing.

Our intercept tells us that when the natural log of weight is 0, the associated natural log of acceleration rate is about 1.75 seconds.

Once again, though, we are going to be looking at the significance of the correlation on our log-log graph as we want to use our linear regression techniques to perform this test. Let's overlay our equation over our log-log graph to take a peek at how our our expected linear relationship fits our data:

```python
plt.scatter(weight_ln, acc_ln, marker = 'o')
plt.plot(weight_ln, weight_ln*wslope + wint)
plt.title("ln(0-60 times) compared to ln(weight)")
plt.xlabel("ln(Weight (kg))")
plt.ylabel("ln(0-60 time (sec))")
```

With all this information we have, we can now begin the process of determining the strength of this correlation. Just based off of eyesight it seems like there is a pretty poor correlation here, but our numbers never lie.

From the `sp.linregress` method we alread know that R = -0.068, but now we need to determine if that correlation is significant. We use the value $\rho$ to represent our ln(population) correlation coefficient.

$H_0: \rho = 0$

A.K.A: $H_0:$ There is not a significant linear relationship 

$H_a: \rho \ne 0$

A.K.A: $H_a:$ Ther is a significant linear relationship

We already have that P = $0.454$ from our `sp.linregress` function, so all we have left to do is motivate a value for alpha and compare the two. Since we tested at a 0.05 significance level in our previous test, we will use the same alpha to maintain consistency.

```python
rejectCheck(alpha, wpval)
```

That makes things pretty easy for us! We find that we cannot reject our null hypothesis, which means that we can't reject the claim that there is no significant linear correlation between ln(Weight) and ln(Acceleration Times). Further, this means that there is no significant exponential correlation between Weight and Acceleration Times. 

Given that our horsepower vs acceleration data did have a significant correlation, we can say we have sufficient evidence to make the claim that horsepower has a greater correlation with acceleration time than weight does. Given this fact I believe now I can explain why we had a negative correlation between weight and acceleration time: Engines that can output greater horsepower tend to be heavier due to the extra components and size needed to achieve higher horsepower, so I think the reason we see this unexpected negative trend in our ln(Weight) vs ln(Acceleration Times) is because as the size of the engine increases (to accompany greater horsepower) so does the acceleration time and the weight, which would suggest that weight and horsepower are both dependent on the engine size. So, the negative trending we see is due to the fact that as engine size increases, horsepower and car weight both increase as well, and since horsepower correlates more strongly with acceleration times we have essentially established that greater engine sizes mean greater horsepower, greater weight, and faster acceleration times, hence the negative trend in our weight data. However, that claim needs to be investigated further, at the moment it is just how I am interpreting this trend in the data.
