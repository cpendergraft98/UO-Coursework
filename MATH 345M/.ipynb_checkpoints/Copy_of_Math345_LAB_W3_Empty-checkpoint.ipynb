**Gaussian Distribution**

This week we will go over some basic manipulations and statistical analysis of a Gaussian Distribution.

To remind everyone, a Gaussian probability density function, commonly expressed and $\mathcal{N}(\mu, \sigma)$ has the following formula:

$$
P(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{\frac{(x-\mu)^2}{2\sigma^2}}.
$$

A common and useful special case of a gaussian distribution is the standard normal distribution. A standard normal distribution has mean $\mu = 0$ and variance $\sigma = 1$,  $\mathcal{N}(0,1)$.

```python
# Normal Distribution

import numpy as np
from scipy.stats import norm       #cdf
from scipy.integrate import quad   #integrate

import matplotlib.pyplot as plt
```

To generate a standard normal distribution one could use the `random.randn(d0, d1, ...)` function from the `numpy` library, where `d0`, `d1`, `...` , are the list dimensions.

```python
# let's generate an array of size ( 100, 5 ) filled with values from a standard normal

stand_norm_ar = np.random.randn(100, 5)

radom_row_index  =  np.random.randint(0,100)

print(stand_norm_ar[radom_row_index])
```

```python
# generate a gaussian dataset - standard normal (mean = 0 varience = 1) of a certain size

size = 50000

stand_norm = np.random.randn(size)


print(f'The mean of the dataset is {np.round(stand_norm.mean(),2)} with a standard deviation of {np.round(stand_norm.std(),2)}.\n')

# PLotting Histogram
plt.figure(figsize = (10,10))

plt.hist(stand_norm, bins = 300, color = 'salmon', label = 'stand norm')

plt.legend(fontsize = 15)

plt.title("Standars Normal Histogram\n", fontsize=25)

plt.xlabel("\n x values (%)", fontsize=20)

plt.tick_params(direction='out', length=10)

plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15, rotation=0)

plt.show()

```

```python
# Changing the standars devaiation of the distribution is equivalent to multiplying (spreading) the standars normal by the set amount
st_dev = 0.5

std_dev_dist = st_dev* stand_norm

print(f'The mean of the dataset is {np.round(std_dev_dist.mean(),2)} with a standard deviation of {np.round(std_dev_dist.std(),2)}.\n')

# Plotting the modification

plt.figure(figsize = (10,10))

plt.hist(std_dev_dist, bins = 300, color = 'blue', label = 'new_stdev')
plt.hist(stand_norm, bins = 300, color = 'salmon', alpha = 0.5 , label = 'stand norm')

plt.legend(fontsize = 15)

plt.title("Gaussian Distributions Histogram\n", fontsize=25)

plt.xlabel("\n x values", fontsize=20)

plt.tick_params(direction='out', length=10)

plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15, rotation=0)

plt.show()



# # Changing the mean of the distribution is equivalent to shifting the dataset by the set amount

mean = 5.0

gauss = std_dev_dist + mean

print(f'The mean of the dataset is {np.round(gauss.mean(),2)} with a standard deviation of {np.round(gauss.std(),2)}.\n')

# Plotting the modification

plt.figure(figsize = (10,10))

plt.hist(std_dev_dist, bins = 300, color = 'blue', alpha = 0.5 ,label = 'new_stdev')
plt.hist(stand_norm, bins = 300, color = 'salmon', alpha = 0.5 , label = 'stand norm')
plt.hist(gauss, bins = 300, color = 'green', alpha = 0.5 , label = 'gauss')


plt.legend(fontsize = 15)

plt.title("Gaussian Distributions Histogram\n", fontsize=25)

plt.xlabel("\n x values", fontsize=20)

plt.tick_params(direction='out', length=10)

plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15, rotation=0)

plt.show()

```

**Problem 1:**
In this excercise, you will develop and write down (step by step) an algorithm that will normalize a gaussian distribution of a certain mean $\mu$ and variance $\sigma$. 

Then, you will implement the set algorithm on a generated dataset with a values for  $( \mu , \sigma )$ randomly picked from the ranges $( [0,10] , [0, 50] )$

Plot the histograms of your original and normalized datasets (together in one plot).

Don't forget to *label your axis*, show a *legend*, and *name your plot*!

```python
# Solution:

def normalize_data_set(data_set):
    mu = data_set.mean() # Expected numpy data set
    sigma = data_set.std() 
    return (data_set - mu) / sigma
```

```python
norm_ar = np.random.randn(100000)
sigma = np.random.randint(0, 50)
print("Sigma:", sigma)
mu = np.random.randint(0, 10)
print("Mean:", mu)

gen_data_set = (norm_ar * sigma) + mu

new_array = normalize_data_set(gen_data_set)

plt.figure(figsize = (10,10))

plt.hist(gen_data_set, bins =50, color = 'blue', alpha = 0.5 ,label = 'Non-Normalized')

plt.hist(new_array, bins = 50, color = 'salmon', alpha = 0.5 , label = 'Normalized')

plt.legend(fontsize = 15)

plt.title("Normalized and Non-Normalized Gaussian Distributions\n", fontsize=25)

plt.xlabel("\n x values", fontsize=20)

plt.show()
```

**Section 2:**

Next, we're gonna perform a bit of analysis on a Gausian probability density function. 

For our analysis, we're gonna start with a standard normal case ($\mu = 0$, $\sigma = 1$), however keep in mind that you can normalize any Gaussian distribution through the procedure developed in Problem 1.

$$
P(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{\frac{(x-\mu)^2}{2\sigma^2}}.
$$


```python
# All your function definitions should be presented at the begining of your code, however for this tutorial, 
# we will write it in the middle of the text due to the flow (let's see how many of you move this to the beginnning)

def gaussian(x, mu, sigma):
    f_x = (1 / np.sqrt(2*np.pi*sigma*sigma) ) * np.exp( - (x - mu)*(x - mu) / (2*sigma*sigma) )

    return f_x
```

```python
# Create Gaussian Probability Density Function
low_limit = -100
high_limit = 100
dx = 0.01

x = np.arange( low_limit, high_limit, dx)

mu = 0.0
sigma = 1.0

mu_1 = 3.0
sigma_1 = 2.5

P_x = gaussian(x, mu, sigma)
P_x1 = gaussian(x, mu_1, sigma_1)
```

```python
plt.figure(figsize = (10,10))

plt.plot(x, P_x, color = 'blue', linewidth = 4.0, label = 'Stand Norm')
plt.plot(x, P_x1, color = 'red', linewidth = 4.0, label = 'Gaussian')

plt.xlim(-10,10)

plt.legend(fontsize = 15)

plt.title("Gaussian Probability Density Function\n", fontsize=25)

plt.xlabel("\n x values", fontsize=20)
plt.ylabel("\n P(x) ", fontsize=20)

plt.tick_params(direction='out', length=10)

plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15, rotation=0)

plt.show()
```

For this excercise you will prove the Emperical Rule that you used to check for outliers in *Lab 1* using $\sigma, 2\sigma,$ and $3\sigma$ intervals. The `integrate.quad()` function from the `scipy` library will serve you useful. Explain your results. 

```python
# Solution

def gaussian(x, mu, sigma):
    f_x = (1 / np.sqrt(2*np.pi*sigma*sigma) ) * np.exp( - (x - mu)*(x - mu) / (2*sigma*sigma) )

    return f_x
```

```python
for i in range(5):
    sigma = np.random.randint(1000)
    print("Sigma:", sigma)
    mu = np.random.randint(1000)
    print("Mu:", mu)
    for j in range(1, 4):
        print(f"Amount of the distribution within {j} distributions")
        print(quad(gaussian, mu - (j * sigma), mu + (j*sigma), args = (mu, sigma))[0], "\n")
        
# We can see with random distributions that +/- 1 standard deviation is about 68%, +/- 2 std is about 95%, and +/- 3 std is 
# about 99%, which agrees with the emperical rule.
        
```

Now let's consider the Cummulative Distribution Function (CDF):


\begin{align*}
F_{Z}(z) &= P(Z \leq z) \\
        &= \int_{-\infty}^{z} P(x) dx
\end{align*}

and the Inverse Cumulative Distribution Funcion (ICDF). In of the distribution function $F_Z$, the ICDF $Q(p)$ returns a value $z$ such that:

$$
F_{Z}(z) := P(Z \leq z) = p
$$
