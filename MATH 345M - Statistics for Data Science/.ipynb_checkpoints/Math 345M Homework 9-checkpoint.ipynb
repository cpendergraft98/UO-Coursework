**Problem 12.74:** The following table shows data on average per capita coffee consumption and hear disease rate in a random sample of 10 countries.

|Yearly coffee consumption in liters|Death from heart diseases|
|:----------------------------------|:------------------------|
|2.5|221|
|3.9|167|
|2.9|131|
|2.4|191|
|2.9|220|
|0.8|297|
|9.1| 71|
|2.7|172|
|0.8|211|
|0.7|300|

**a)** Create a scatter plot of the data.

From our data above, we can interpret the amount of coffee consumed as an independent variable which we can plot on the x axis, and the deaths from heart disease as the dependent variable on our y-axis.

```python
x = np.array([2.5, 3.9, 2.9, 2.4, 2.9, 0.8, 9.1, 2.7, 0.8, 0.7])
y = np.array([221, 167, 131, 191, 220, 297, 71, 172, 211, 300])

plt.scatter(x, y, marker = 'o')
plt.title('Deaths from heart disease compared to coffee consumption')
plt.xlabel("Amount of coffee consumed per capita (liters)")
plt.ylabel("Deaths from heart disease")
```

**b)** Find the equation of the least-squares regression line. Add this to your scatter plot from part a.

Scipy.stats provides us with a nice little method `sp.linregress(x, y)` that will allow us to gather the necessary information to construct the equation of our least-squares regression line.

```python
slope, intercept, rvalue, pvalue, stderr = sp.linregress(x, y)
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", rvalue)
print("P-value:", pvalue)
print("Standard error:", stderr)
```

So, based on the above data, the equation of our least-squares regression line is going to be

$y = -23.9x + 266.6$

Now we just need to lay this line onto our scatter plot:

```python
plt.scatter(x, y, marker = 'o')
plt.plot(x, slope*x + intercept)
plt.title('Deaths from heart disease compared to coffee consumption')
plt.xlabel("Amount of coffee consumed per capita (liters)")
plt.ylabel("Deaths from heart disease")
```

**c)** Explain in words what the slope and y-intercept of the regression line tell us.

So we know from the regression line that we have a slope of about -23.9. Given what we know about rise over run, this tells us that as the coffee consumption per capita goes up by 1, the associated death rate is expected to decrease by a factor of 23.9. The y-intercept tells us that when there is absolutely no coffee consumption per capita we can expect the fixed number of deaths due to heart disease to be about 266.6.

**d)** How well does the regression line fit the data? Explain your response.

So, we already have that our correlation coefficient is R = 0.836 from our `sp.linregress(x, y)` method, and just eyeballing that value I think it is safe to say that we have a pretty strong linear correlation here given how close 0.836 is to 1. However, this is a statistics course, so we should always make sure to do our due dilligence and make sure that we are getting appropriate measurements. I am going to go off the books methodology for this where I will use a hypothesis test to determine if the corellation coefficient is significant.

We use the value $\rho$ to represent our population correlation coefficient, which is unknown. If $\rho$ = 0, then we can determine that there is no significant linear relationshipt between x and y. However, if $\rho \ne 0$, then there IS some significant linear relationship between x and y. Based on the information in our book, we should have the following null and alternative hypotheses:

$H_0: \rho = 0$

A.K.A: $H_0:$ There is not a significant linear relationship 

$H_a: \rho \ne 0$

A.K.A: $H_a:$ Ther is a significant linear relationship

The textbook typically suggests that we use a significance level of 0.05 to perform our hypothesis testing, but given that in this situation we are looking at deaths from heart disease and how that relates to coffee consumption per capita, I am inclined to instead use a significance level of 0.01 for this test.

We have the p-value from our previous use of the `sp.linregress(x, y)` method, so all we need to do now is compare our $\rho$ and $\alpha$ values in order to decide if our null hypothesis should be rejected or not.

```python
def rejectCheck(alpha, p):
    # Checks if a null hypothesis should be rejected. A return of 'True' indicates H_0 should be rejected.
    # A return of 'False' indicates H_0 should not be rejected.
    return alpha > p

alpha = 0.01
rejectCheck(alpha, pvalue)
```

Okay! So we find that we have enough evidence to reject our null hypothesis that the population correlation coefficient is not significantly different from zero. This leads us to accept our alternative hypothesis, which is that there is a significant linear relationship between x and y!

**e)** Which point has the largest residual? Explain what the residual means in context. Is this point an outlier? An influential point?

So to compute our residual all we need to do is take our measured y-values and subtract them from the expected y-values from our regression line, and the resulting difference with the largest absolute value will represent our largest residual, which is the largest vertical distance between the regression line and the measured point.

First, I need to create a data set of the expected y-values from our regression line:

```python
y_exp = []
for i in x:
    val = slope*i + intercept
    y_exp.append(val)
    
print(y_exp)
```

Now, all we have to do is subtract each element at position i in our expected list from the element at position i in our measured y value list, populate a new list with the residals, and then determine which among them has the highest residual:

```python
residual_list = []
for i in range(len(y_exp)):
    res = y[i] - y_exp[i]
    residual_list.append(res)
    
print(residual_list)
```

So just looking at this data, we see that our greatest residual is going to be 66.38, which rests at position [2] within our list of residuals. This means it corresponds to the measured point (2.9, 131). This means that when our coffee consumption per capita is at 2.9 liters and our observed deaths are 131 it is about 66 less than the predicted value.

In order to determine if this is an influential point or just an outlier, we should try removing it from our plot and checking the resulting graph. If we have a drastically different slope then we would say that our point is definitely influential, as its removal changed our regression line. However, if the point is just an outlier, we shouldn't really see that drastic of difference between the two slopes.

```python
x_resremoved = np.array([2.5, 3.9, 2.4, 2.9, 0.8, 9.1, 2.7, 0.8, 0.7])
y_resremoved = np.array([221, 167, 191, 220, 297, 71, 172, 211, 300])

rrslope, rrintercept, rrrvalue, rrpvalue, rrstderr = sp.linregress(x_resremoved, y_resremoved)
print("Slope:", rrslope)
print("Intercept:", rrintercept)
print("R-value:", rrrvalue)
print("P-value:", rrpvalue)
print("Standard error:", rrstderr)
```

So without our residual, our slope changes from about -23.9 to about -23.8. This is certainly not a drastic change, which leads me to believe that the point (2.9, 131) is probably just an outlier.

**d)** Do the data provide convincing evidence that there is alinear relationship between population size and homicide rate? Carry out an appropriate test at a significance level of 0.05 to help answer this question.

So now we just need perform another relatively simple hypothesis test. Once again we use $\rho$ to represent our population correlation, and have the following null and alternative hypotheses:
